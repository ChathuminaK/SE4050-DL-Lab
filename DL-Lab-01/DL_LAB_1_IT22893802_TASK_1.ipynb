```python
import numpy as np

import numpy as np

# Create a random 4x4 array from an exponential distribution
# The 'scale' parameter is the inverse of the rate parameter (lambda). Default is 1.0.
# The 'size' parameter defines the output shape.
exponential_array_4x4 = np.random.exponential(scale=1.0, size=(4, 4))

print("Random 4x4 Array from Exponential Distribution:")
print(exponential_array_4x4)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data from three different distributions
uniform_data = np.random.rand(100000)
normal_data = np.random.randn(100000)
exponential_data = np.random.exponential(size=100000)

# Plotting the histograms
# Using 200 bins for a smoother look
plt.hist(uniform_data, density=True, bins=200, histtype="step", color="blue", label="Uniform (rand)")
plt.hist(normal_data, density=True, bins=200, histtype="step", color="red", label="Normal (randn)")
plt.hist(exponential_data, density=True, bins=200, histtype="step", color="green", label="Exponential")

# Adjusting the plot view for better visualization
plt.axis([-2.5, 5, 0, 1.1]) # Changed x-axis limit to 5 to see the exponential tail
plt.legend(loc="upper right") # Changed loc to "upper right" for better placement
plt.title("Comparison of Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Define the range for x and y
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)

# Create a meshgrid
X, Y = np.meshgrid(x, y)

# Calculate Z
Z = X**2 + Y**2

# Plotting the 3D surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title("$Z = X^2 + Y^2$")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")

plt.show()
```

```python
# This code should be run inside the seaborn_tutorial.ipynb notebook

# Assuming the DataFrame is loaded as 'df'
# (The tutorial likely uses a Pokemon dataset)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Select the specified features
features_df = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]

# 2. Calculate Pearson correlation
pearson_corr = features_df.corr(method='pearson')

# 3. Visualize Pearson correlation with a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Pearson Standard Coefficient")
plt.show()

# 4. Calculate Spearman correlation
spearman_corr = features_df.corr(method='spearman')

# 5. Visualize Spearman correlation with a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Spearman Rank Correlation")
plt.show()
```
